2. Starbucks Customer Profile

```python
profile=pd.read_csv('profile.csv')
profile.head()
```

```python
profile.info()
#Some values from gender and income are missing
```

```python
#delete column "Unnamed: 0"
profile.drop(columns='Unnamed: 0', inplace=True)
```

```python
profile.describe()
```

```python
# check missing values
profile.isna().sum()
```

```python
# check if the missing gender and income are from the same rows
missing_gender = profile[profile['gender'].isna()]
missing_income = profile[profile['income'].isna()]

np.sum(missing_gender['id'] == missing_income['id'])
```

```python
# They are from the same rows. Since there are 17000 records in total, the 2715 records do not provide the important information for the profile, and filling the gap with the median value will skew the anaylis,
# We can delete the 2175 records instead
new_profile = profile[~profile['id'].isin(missing_gender['id'])].reset_index()
new_profile.info()
```

```python
#change column 'became_member_on" from int to date time
new_profile['became_member_on'] = pd.to_datetime(new_profile['became_member_on'], format = '%Y%m%d')
```

```python
#there are also some extremly high aged values, the highest is 118, which should not happen
#check how many records are over 90 years old
new_profile[(new_profile['age'] < 119) & (new_profile['age'] > 90)]
```

```python
#removing the records does not really hurt the anaysis as the data size is still big enough
new_profile.drop(new_profile.index[(new_profile['age'] < 119) & (new_profile['age'] > 90)], inplace = True)
```

```python
#Save the new profile into csv
new_profile.to_csv('cleaned_profile.csv', index=False)
```

4. Starbucks Customer Transcript

```python
transcript = pd.read_csv('transcript.csv')
transcript.head()
```

```python
#drop column 'Unnamed: 0', and change the column name 'person' into 'id'
transcript.drop(columns='Unnamed: 0', inplace=True)
transcript.rename(columns={'person': 'id'}, inplace=True)
```

```python
#extract the offer id from the 'value' column, and rename the column as 'offer_id'
import json
# Convert string representations to dictionaries
transcript['value'] = transcript['value'].apply(lambda x: json.loads(x.replace("'", "\"")))
transcript['offer_id'] = transcript['value'].apply(lambda x: x.get('offer id'))
```

```python
#drop the column 'value'
transcript.drop(columns= ['value'], inplace=True)
```

```python
transcript.head()
```

```python
transcript.describe()
```

```python
#save the data into a new csv
transcript.to_csv('cleaned_transcript.csv', index=False)
```

5. Starbucks Store Locations 2021

```python
store_21 = pd.read_csv('Starbucks Store Locations 2021.csv')
store_21.head()
```

```python
#drop unnecessary columns
store_21.drop(columns=['Unnamed: 0', "schedule", 'slug', 'streetAddressLine2', 'streetAddressLine3', 'postalCode','currentTimeOffset', 'windowsTimeZoneId'], inplace=True)
```

```python
store_21.rename(columns={'storeNumber':'Store Number', 'countryCode': 'Country', 'ownershipTypeCode':'Ownership Type', 'latitude':'Latitude', 'longitude':'Longitude', 'streetAddressLine1': 'Street Address', 'city':'City', 'countrySubdivisionCode': 'State/Province', 'olsonTimeZoneId': 'Timezone'}, inplace=True)
```

```python
store_21
```

```python
#the language of the city is mixed
store_21.loc[store_21['City'] == '香港', 'City'] = 'Hong Kong'
store_21.loc[store_21['City'] == 'Kowloon', 'City'] = 'Hong Kong'
```

```python
store_21.info()
```

```python
store_21['Ownership Type'].value_counts()
```

```python
#replace the ownership types by Company Owned, Licensed, Joint Venture,Franchise
store_21.loc[store_21['Ownership Type'] == 'CO', 'Ownership Type'] = 'Company Owned'
store_21.loc[store_21['Ownership Type'] == 'LS', 'Ownership Type'] = 'Licensed'
store_21.loc[store_21['Ownership Type'] == 'JV', 'Ownership Type'] = 'Joint Venture'
store_21.loc[store_21['Ownership Type'] == 'FR', 'Ownership Type'] = 'Franchise'
```

```python
store_21.head()
```

```python
#Add a column of year
store_21['Year'] = 2021
```

```python
store_21.to_csv('starbucks_2021.csv', index=False)
```

```python
store_21[store_21['Country'] == 'CN']
```
